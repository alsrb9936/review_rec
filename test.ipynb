```python
import torch
import torch.nn as nn

def glove_load_embedding(path):
    with open(path, encoding='utf-8') as f:
        word_emb = list()
        word_dict = dict()
        word_emb.append([0])
        word_dict['<UNK>'] = 0
        for line in f.readlines():
            tokens = line.split(' ')
            word_emb.append([float(i) for i in tokens[1:]])
            word_dict[tokens[0]] = len(word_dict)
        word_emb[0] = [0] * len(word_emb[1])
    return word_emb, word_dict

word_emb, word_dict = glove_load_embedding("/home/infolab/mnt/mingyu/review_rec/review_reproducibility/word2vec/glove.6B.50d.txt")
embedding = nn.Embedding.from_pretrained(torch.Tensor(word_emb))
```
